# 年度分析

获取月线数据，排序。

```python
df = ts.get_hist_data('300036',ktype='M')
```

```python
df.head()
```

```python
len(df)
```

```python
# 绘制价格曲线图。
def splot(dfx):
    fig=plt.figure(figsize=(48,4), dpi=80)
    p1=fig.add_subplot(1,1,1)

    p1.plot(dfx.index,dfx['open'])#x轴y轴
    p1.plot(dfx.index,dfx['close'])
    p1.plot(dfx.index,dfx['high'])
    p1.plot(dfx.index,dfx['low'])
    plt.grid(True)
    plt.gcf().autofmt_xdate()#？？
    plt.show()

# 绘制价格变动率和变动的百分比率。    
def splotbar(dfx):
    fig=plt.figure(figsize=(48,4), dpi=80)
    p1=fig.add_subplot(1,1,1)

    p1.plot(dfx.index,dfx['p_change'], color='red')#线形图
    p1.bar(dfx.index,dfx['price_change'])#柱行图
    plt.grid(True)
    plt.gcf().autofmt_xdate()
    plt.show()
```

```python
# 排序
df2=df.sort_index(ascending=True)
```

### 绘制价格曲线图，价格变动与变动率曲线图。

```python
splot(df2)
```

```python
splotbar(df2)
```

### 变动率分析。

```python
# 获取价格变动及其变动率，构建新的df2v，计算累积求和，并绘制曲线。
df2v = pd.DataFrame(df2,columns=("price_change","p_change")).sort_index(ascending=True)#升序排列
df2c=df2v.cumsum()

df2c.plot(kind='line',figsize=(32,4),grid=True)
df2c.plot(kind='bar',figsize=(32,4),grid=True)
```

# 逐年分析
获取周线数据分析绘图，进行逐年比较。

```python
def getYear(year=2018):#取数
    dfy = ts.get_hist_data('300036', start=str(year)+"-01-01", end=str(year+1)+"-01-01", ktype='W')
    dfy = dfy.sort_index(ascending=True)
    return dfy

def plotYear(year=2018):#画图
    dfy = getYear(year)
    splot(dfy)#怎么知道只画前4列数值？
    splotbar(dfy)##怎么知道画某列数值？
    return dfy
```
