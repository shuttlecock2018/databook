# 股票分类数据

- 来自 http://tushare.org/classifying.html 的股票分类数据。

分类数据提供股票的分类信息数据，从股票类型的不同角度进行数据分类，在一些统计套利方法的应用中，时常会以股票的分类来做切入，比如对某些行业或概念进行阶段统计来决定近期的交易策略等。TuShare提供的分类数据主要包括以下类别：

    行业分类
    概念分类
    地域分类
    中小板分类
    创业板分类
    风险警示板分类
    沪深300成份股及权重
    上证50成份股
    中证500成份股
    终止上市股票列表
    暂停上市股票列表

行业分类

在现实交易中，经常会按行业统计股票的涨跌幅或资金进出，本接口按照sina财经对沪深股票进行的行业分类，返回所有股票所属行业的信息。考虑到是一次性在线获取数据，调用接口时会有一定的延时，请在数据返回后自行将数据进行及时存储。sina财经提供的行业分类信息大致如下图所示：

返回值说明：

    code：股票代码
    name：股票名称
    c_name：行业名称

调用方法：

import tushare as ts

ts.get_industry_classified()



````python
import tushare as ts

ts.get_industry_classified()
````

# 地域分类

按地域对股票进行分类，即查找出哪些股票属于哪个省份。

参数说明：

    file_path:文件路径，默认为None即由TuShare提供，可以设定自己的股票文件路径。

返回值说明：

    code：股票代码
    name：股票名称
    area：地域名称

调用方法：

import tushare as ts

ts.get_area_classified()



````python
import tushare as ts

ts.get_area_classified()
````

````python
#beijing.columns 可以列出列索引名称

import tushare as ts
beijing=ts.get_area_classified()

#beijing[beijing['area']=="北京"]#列出area是北京的,必须【】，（）不可以，未写=即未定义，只是临时赋值

beijing=beijing[beijing['area']=="北京"]#=是正式定义函数，下一个前20行就是此函数的

beijing[:20]
````
