### 使用pil影像处理库。

```python
from PIL import Image, ImageFilter
```

```python
# 自定义的影像显示，使用HTML标签。刷新好像不更新！
def showimg(imgurl,alt=""):
    display(HTML('<img src='+ "'"+imgurl+"'"+ ' align=left alt=alt />'))
```

```python
#Read image
im = Image.open( 'tulip.jpg' )
```

```python
#Viewing EXIF data embedded in image
exif_data = im._getexif()

exif_data
```

```python
#Display image
im
```

```python
im_blur = im.filter( ImageFilter.BLUR )
im_blur.save( 'image_blur.jpg', 'JPEG' )
im_blur
```

```python
#Applying a filter to the image
im_sharp = im.filter( ImageFilter.SHARPEN ).filter( ImageFilter.SHARPEN ).filter( ImageFilter.SHARPEN )

#Saving the filtered image to a new file
im_sharp.save( 'image_sharpened.jpg', 'JPEG' )
im_sharp
```

```python
#Splitting the image into its respective bands, i.e. Red, Green,
#and Blue for RGB
r,g,b = im_sharp.split()
```

```python
#im.histogram()

import PIL
from PIL import Image
from matplotlib import pyplot as plt

im = Image.open('tulip.jpg')  
w, h = im.size  
colors = im.getcolors(w*h)

def hexencode(rgb):
    r=rgb[0]
    g=rgb[1]
    b=rgb[2]
    return '#%02x%02x%02x' % (r,g,b)

for idx, c in enumerate(colors):
    plt.bar(idx, c[0], color=hexencode(c[1]))

plt.show()
```
